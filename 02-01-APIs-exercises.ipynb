#### Exercise 1-A. Extract the first name, last name, and team of 200 NBA players. Return the results in a DataFrame.

```python
# for dealing with data
import pandas as pd

# for using APIs
import requests

# for data processing
import re

def ballondontlie_api_call(url, params):
    
    headers = {"Authorization": "4ce19b34-7633-42ff-8521-7cf18d4196db"}
    response = requests.get (url, headers = headers, params = params)

    print(f"\nAPI call response type: {response.status_code}\n")
    
    if response.status_code == 200:
        data = response.json()
        print (f"\nENDPOINT RESPONSE:\nData type: {type(data)}\nLength: {len(data)}\nKeys: {data.keys()}\nFull Data:")
        print(data)
        return data

    else:
        print(f"API call failed due to: {response.reason}")


# create one list for each of the columns needed: name, last name, team
first_name = []
last_name = []
team = []

for i in range(0, 200, 25):
    params = {"cursor":i}
    data = ballondontlie_api_call ('https://api.balldontlie.io/v1/players', params = params)
    players_data = data.get('data') # accesing players data (first element of the dictionary received as the endpoint response)

    # iterate the list that contains all player data and store needed data in the correspondent list 

    for player in players_data:
        first_name.append(player.get("first_name"))
        last_name.append(player.get("last_name"))
        team.append(player.get("team").get("full_name"))

#printing the lists (for testing only) 
print("\nFIRST_NAME LIST\n")
print(first_name)
print("\nLAST_NAME LIST\n")
print(last_name)
print("\nTEAM LIST\n")
print(team)

print("\nPLAYERS INFO AS DATAFRAME\n")
players_info = {"first_name":first_name, "last_name":last_name, "team":team}

players_info_df = pd.DataFrame(players_info)

print(players_info_df)
```
