````python
# 1. Usando la siguiente lista: 

# temperaturas = [2,45,67,23, 45,66,12,1,34,67,99,98,67,64, 44, 100]

#Crea tres listas nuevas, donde: 

#- En la primera tengamos temperaturas entre 0 - 10, llama a la nueva lista `bajas_temp`.

#- En la segunda tengamos temperaturas entre 11- 20, llama a la nueva lista `medias_temp`.

#- En la tercera tengamos temperaturas por encima de 20, llama a la nueva lista `altas_temp`


temperaturas = [2,45,67,23, 45,66,12,1,34,67,99,98,67,64,44,100]

bajas_temp = []
medias_temp = []
altas_temp = []

for temperatura in temperaturas:
    if 0 <= temperatura <= 10:
        bajas_temp.append(temperatura)
    elif 11 < temperatura <= 20:
        medias_temp.append(temperatura)
    elif temperatura > 20:
        altas_temp.append(temperatura)

print (f"la sub-lista de temperaturas bajas contiene: {bajas_temp}")
print (f"la sub-lista de temperaturas medias contiene: {medias_temp}")
print (f"la sub-lista de temperaturas altas contiene: {altas_temp}")
````

````python
# 2. Las temperaturas están grados centígrados. Para cada una de las listas creadas en el ejercicio anterior, convertid las temperaturas a grados fahrenheit

# la relacion de grados celsius y fahrenheit es:  (0 °C × 9 / 5) + 32 = °F

# resolución  con list comprehention

bajas_temp_f = [round(temperatura * (9/5) + 32, 1) for temperatura in bajas_temp]

print(f"las lista de temperaturas bajas en fahrenheit contiene: {bajas_temp_f}")

medias_temp_f = [round(temperatura * (9/5) + 32, 1) for temperatura in medias_temp]

print(f"las lista de temperaturas bajas en fahrenheit contiene: {medias_temp_f}")

altas_temp_f = [round(temperatura * (9/5) + 32, 1) for temperatura in altas_temp]

print(f"las lista de temperaturas bajas en fahrenheit contiene: {altas_temp_f}")


# resolucion con for loop

bajas_temp_f_1 = []

for temperatura in bajas_temp:
    temp_f = round (temperatura * (9/5) + 32, 2)
    bajas_temp_f_1.append (temp_f)

print(bajas_temp_f_1)

#-------

medias_temp_f_1 = []

for temperatura in medias_temp:
    temp_f = round (temperatura * (9/5) + 32, 2)
    medias_temp_f_1.append(temp_f)

print(medias_temp_f_1)

#------

altas_temp_f_1 = []

for temperatura in altas_temp:
    temp_f = round (temperatura * (9/5) + 32, 2)
    altas_temp_f_1.append(temp_f)

print(altas_temp_f_1)

````
